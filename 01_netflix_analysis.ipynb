```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = r"C:\Users\sanjay pallav\Documents\Python banny\Pandas_Mini_Projects\datasets\netflix_titles.csv"

# Load the dataset
data = pd.read_csv(file_path, encoding='utf-8')

#Display first few rows to verify
print(data.head())
```

```python
# Cleaning all the null values in the dataset by dropping them
data.dropna(inplace=True)

# Checking if all the null values are cleaned or not.
data.isnull().sum()
```

```python
# Checking how many movies vs Tv shows using value
data["type"].value_counts()
```

```python
# Listing out the top 10 directors with most movie titles
data["director"].value_counts().head(10)
```

```python
# Listing out the number of titles per year
data["release_year"].value_counts().sort_index()
```

```python
# Optional: Method 2 using loop and Matplotlib (messy but exploratory)
# This was just an experiment to try visualizing via manual looping

for years in data["release_year"].unique():
    count = data[data["release_year"]==years]["title"]
    plt.plot(count)
    plt.show()
```

```python
# Optional: Method 3 - Clean  one-liner with Matplotlib
plt.plot(data["release_year"].unique(), [data[data["release_year"] == y]["title"].count() for y in data["release_year"].unique()], '-o')
```

```python
# Filter:Titles from India after 2015 with rating 'TV-14' or 'R'.
data[(data["country"]=="India")&(data["release_year"]>2015)&(data["rating"].isin(["TV-14", "R"]))]
```
